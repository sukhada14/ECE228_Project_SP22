```python
mean_spectral_values = {}
for crop in crop_types:
    df = df_metadata[df_metadata['cropType'] == crop]
    spectral_bands_13 = []
    spectral_bands_last = []
    
    for i, row in df.iterrows():
        spectral_bands_13.append(row['Input_Data'].data[:13, :])
        spectral_bands_last.append(row['Input_Data'].data[13, :])
    
    spectral_bands_13 = np.array(spectral_bands_13)
    spectral_bands_last = np.array(spectral_bands_last)

    mean_spectral_values[crop] = {}
    mean_spectral_values[crop]['13'] = np.mean(spectral_bands_13, axis = 0)
    mean_spectral_values[crop]['14'] = np.mean(spectral_bands_last, axis =0)
```

```python
list(mean_spectral_values['Maize'].values())[0].shape
```

```python
def get_x_y_13bands(crop_name):
    xs = []
    ys = []

    for ts in mean_spectral_values[crop_name]['13']:
        x = []
        y = []
        for i, val in enumerate(ts):
            if val != 0:
                x.append(i)
                y.append(val)
        xs.append(x)
        ys.append(y)
        
    return xs, ys
```

```python
def get_x_y_14band(crop_name):
    x = []
    y = []
    for i, val in enumerate(mean_spectral_values[crop_name]['14']):

        if val != 0:
            x.append(i)
            y.append(val)
    return x, y
```

```python
spectral_band_dict = \
{
    1: 'AEROS',
    2: 'BLUE',
    3: 'GREEN',
    4: 'RED',
    5: 'RDED1',
    6: 'RDED2',
    7: 'RDED3',
    8: 'NIR',
    9: 'RDED4',
    10: 'VAPOR',
    11: 'CIRRU',
    12: 'SWIR1',
    13: 'SWIR2'
    
}
```

```python
plt.rcParams.update({'font.size': 10})
```

```python
def show_13_bands(xs, ys, crop_name):
    
    figure(figsize=(8, 6), dpi=80)
    for i in range(1, 14):
        plt.plot(xs[i-1], ys[i-1], label = spectral_band_dict[i])
        #plt.plot(mean_spectral_values['Maize']['13'][1], label = "line 2")
    
    plt.xticks(fontsize=12)
    plt.yticks(fontsize=12)
    
    plt.legend(bbox_to_anchor=(0.5, -0.44), loc='lower center', ncol =3, fontsize= 12)
    plt.title('Average Spectral band (1-13) variation for '+ crop_name, fontsize= 13)
    plt.xlabel('Day of the year', fontsize= 13)
    plt.ylabel('Average pixel value', fontsize= 13)
    plt.show()
```

```python
def show_14_band(xs, ys, crop_name):
    
    figure(figsize=(8, 6), dpi=80)

    plt.plot(x, y,label = "Spectral Band 14: Green Chlorophyll Vegetation Index")
        #plt.plot(mean_spectral_values['Maize']['13'][1], label = "line 2")
    plt.plot(x, y, 'bo', markersize=4) 
    plt.legend( loc='lower center', fontsize= 12)
    plt.title('Average Spectral band 14 variation for '+crop_name, fontsize=13)
    plt.xlabel('Day of the year', fontsize= 13)
    plt.ylabel('Average GCVI value', fontsize=13)
    plt.xticks(fontsize=12)
    plt.yticks(fontsize=12)
    
    plt.show()
```

```python
for crop in crop_types:
    print('For', crop, ':')
    xs, ys = get_x_y_13bands(crop)
    x, y = get_x_y_14band(crop)
    show_13_bands(xs, ys, crop)
    show_14_band(x, y, crop)
    print('-'*80)
    print('-'*80)
    print()
```

## Model Training on 14th band

```python
from sklearn.datasets import load_iris
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
X = [] 
Y = []

for i, row in df_metadata.iterrows():
    
    X.append(row['Input_Data'].data[13, :])
    Y.append(row['cropType'])

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
def acc(y_pred, y_test):
    count = 0
    for x,y in zip(y_pred, y_test):
        if x == y:
            count +=1
            
    return count/len(y_pred)
```

```python
model = LogisticRegression().fit(X_train, y_train)

y_pred = model.predict(X_test)

accuracy = acc(y_pred, y_test)
accuracy
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import make_classification

clf_rf = RandomForestClassifier(max_depth=20, random_state=0)
clf_rf.fit(X_train, y_train)

y_pred_rf = clf_rf.predict(X_test)

accuracy_rf = acc(y_pred_rf, y_test)
accuracy_rf
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC

clf_svm = make_pipeline(StandardScaler(), SVC(gamma='auto'))
clf_svm.fit(X_train, y_train)

y_pred_svm = clf_svm.predict(X_test)

accuracy_svm = acc(y_pred_svm, y_test)
accuracy_svm
```

```python
from sklearn.neural_network import MLPClassifier

clf_mlp = MLPClassifier(random_state=1, max_iter=1000).fit(X_train, y_train)

y_pred_mlp = clf_mlp.predict(X_test)
accuracy_mlp = acc(y_pred_mlp, y_test)
accuracy_mlp
```

```python
from sklearn.ensemble import GradientBoostingClassifier

clf_gb = GradientBoostingClassifier().fit(X_train, y_train)

y_pred_gb = clf_gb.predict(X_test)
accuracy_gb = acc(y_pred_gb, y_test)
accuracy_gb
```
